# Required Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_digits, load_breast_cancer, load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

# 2d Scatter Plot Function

```python
def plot_2d_scatter(X, y, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
```

# KNN Classification Function

```python
def knn_classification(X_train, X_test, y_train, y_test, k):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy
```

# Breast Cancer Data Set

```python
breast_cancer = load_breast_cancer()
X_bc = breast_cancer.data[:, :2]
y_bc = breast_cancer.target
X_bc_train, X_bc_test, y_bc_train, y_bc_test = train_test_split(X_bc, y_bc, test_size=0.2, random_state=42)

plot_2d_scatter(X_bc, y_bc, 'Breast Cancer Wisconsin (Diagnostic) Dataset - First 2 Features')

k = 5
accuracy_bc = knn_classification(X_bc_train, X_bc_test, y_bc_train, y_bc_test, k)
print("Accuracy on Breast Cancer dataset:", accuracy_bc)
```

# Wine Data Set

```python
wine = load_wine()
X_wine = wine.data[:, :2]
y_wine = wine.target
X_wine_train, X_wine_test, y_wine_train, y_wine_test = train_test_split(X_wine, y_wine, test_size=0.2, random_state=42)

plot_2d_scatter(X_wine, y_wine, 'Wine Dataset - First 2 Features')

k = 5
accuracy_wine = knn_classification(X_wine_train, X_wine_test, y_wine_train, y_wine_test, k)
print("Accuracy on Wine dataset:", accuracy_wine)
```

```python
iris = load_iris()
X_iris = iris.data[:, :2]
y_iris = iris.target
X_iris_train, X_iris_test, y_iris_train, y_iris_test = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

plot_2d_scatter(X_iris, y_iris, 'Iris Dataset - First 2 Features')

k = 5
accuracy_iris = knn_classification(X_iris_train, X_iris_test, y_iris_train, y_iris_test, k)
print("Accuracy on Iris dataset:", accuracy_iris)
```
