# Required Dependencies

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
```

# Load DataSet into the Data Frame

```python
iris = load_iris()
X = iris.data
y = iris.target 

iris_df = pd.DataFrame(X, columns=iris.feature_names)
```

# Specify Attributes

```python

attribute1 = 'sepal length (cm)'
attribute2 = 'sepal width (cm)'
attribute3 = 'petal length (cm)'
```

# Creating Three Dimensional Scatter Plot

```python

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for species, color in zip(range(3), ['blue', 'green', 'red']):
    ax.scatter(iris_df[iris.target == species][attribute1], 
               iris_df[iris.target == species][attribute2], 
               iris_df[iris.target == species][attribute3], 
               c=color, 
               label=iris.target_names[species])

ax.set_xlabel(attribute1)
ax.set_ylabel(attribute2)
ax.set_zlabel(attribute3)
ax.set_title(f'{attribute1} vs {attribute2} vs {attribute3}')

ax.legend()

plt.show()
```
